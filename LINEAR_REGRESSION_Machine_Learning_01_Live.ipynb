```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv("/content/homeprices (2).csv")
```

```python
df
```

```python
plt.scatter(df['area'], df['price'] , color ='red', marker = "+")
plt.xlabel("Area")
plt.ylabel("price")

```

```python
line = LinearRegression()
line.fit(df[['area']], df['price']) #independent var - [[]], deendent var - > []
```

```python
#best m ->  slope, coefficent , coef
line.coef_
```

```python
#best c ->  intercept
line.intercept_
```

```python
line.predict([[3800]]) # how we can make prediction
```

```python
# y_predict = mx + c
#price_pred = m*area + c
135.78767123 * 3800 + 180616.43835616432
```

```python
plt.scatter(df['area'], df['price'] , color ='red', marker = "+", )
plt.plot(df['area'], line.predict(df[['area']]))
plt.xlabel("Area")
plt.ylabel("price")
```
